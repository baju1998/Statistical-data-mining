```python
outlier_vars = [
    "CrimeRate", "SquareFootage", "BackyardSpace", "DistanceToCityCenter", "PreviousSalePrice", "PreviousSalePrice" , "Windows"
]



for var in outlier_vars:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    upperbound = Q3 + 1.5 * IQR
    lowerbound = Q1 - 1.5 * IQR
    print(f"{var} = {upperbound}")
    df[var] = df[var].clip(upper=upperbound,lower=lowerbound)
    #clip upperbound/lowerbound based on outliers in boxplot
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,classification_report, ConfusionMatrixDisplay
import statsmodels.api as sm
import seaborn as sns

variab = [
    "CrimeRate", "SchoolRating", "EmploymentRate", "LocalAmenities",
    "TransportAccess", "SquareFootage", "BackyardSpace",
    "RenovationQualityRate", "DistanceToCityCenter",
    "PropertyTaxRate", "NumBedrooms", "PreviousSalePrice", "AgeOfHome"]


X = df1[variab] #Variables in dataframe to be used

y = df1["IsLuxury"] #dependent variable


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y) #splitting data into 80% train 20% test 
#and ensures same sequence for random number generated
X_train = sm.add_constant(X_train, has_constant="add")#add constant to X variables to have staing point for data
X_test  = sm.add_constant(X_test,  has_constant="add")
model = sm.Logit(y_train, X_train) #creating a model to train using logistic regression
results = model.fit(disp=False)
# Print Model summary
##
##print(results.summary())

y_pred_proba = results.predict(X_test) # using predict that will use coefficients in results to predict if it is Lurxury
y_pred = (y_pred_proba >= 0.5).astype(int) # Convert probabilities to binary predictions

"""print("\nAccuracy Score on Test Data:", accuracy_score(y_test, y_pred))
print("\nClassification Report on Test Data:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix on Test Data:")
print(confusion_matrix(y_test, y_pred))"""

##print(results.summary())
print("Model Summary:")
print(f"This is AIC {results.aic}")
print(f"This is BIC {results.bic}")
print(results.summary())
```

```python
X_train.to_csv("X-Training.csv", index = False)
X_test.to_csv("X-Test.csv", index = False)
y_train.to_csv("y-Training.csv", index = False)
y_test.to_csv("y-Test.csv", index = False)
```

```python
def backward_elimination(X, y, significance_level=0.05): #method passes with listed parameters
    X_opt = X.copy() 
    while True:
        temp_model = sm.Logit(y, X_opt).fit(disp=False)
        p_values = temp_model.pvalues.drop('const', errors="ignore")
        if(len(p_values)) == 0:
            break
        max_p_value = p_values.max()
        if max_p_value >significance_level:
            variable_to_remove = p_values.idxmax()
            print(f"Removing variable: {variable_to_remove} (P-value: {max_p_value:.4f})")
            X_opt = X_opt.drop(columns=[variable_to_remove])
        else:
             
            break
    
    optimized_model = sm.Logit(y, X_opt).fit(disp=False)
    return optimized_model, X_opt


optimized_model, X_train_optimized = backward_elimination(X_train, y_train)

X_test_optimized = X_test[X_train_optimized.columns]

y_pred_proba = optimized_model.predict(X_test_optimized)
y_pred = (y_pred_proba >= 0.5).astype(int)

##print(f" Here {model.params}")
print("\nFinal Model Summary:")
print(f"This is AIC {optimized_model.aic}")
print(f"This is BIC {optimized_model.bic}")
print(optimized_model.summary())




predictors = optimized_model.model.exog_names
eq = ""
for name, coef in optimized_model.params.items():
   if name == "const":
       eq +=f"{coef:.7f}"
   else:
       eq +=f"+ {coef:.7f}({name}) "
print(eq)




print("\nAccuracy Score on Test Data:", accuracy_score(y_test, y_pred))
print("\nClassification Report on Test Data:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix on Test Data:")
cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
disp = ConfusionMatrixDisplay(cm, display_labels=["Not Luxury (0)", "Luxury (1)"])
disp.plot()
plt.show()
```
